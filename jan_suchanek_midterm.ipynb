Firstly, I will import few necessary things:

```python
from urllib.request import urlopen as uReq 
from bs4 import BeautifulSoup as soup
import pandas as pd
from IPython.display import display, HTML
```

Now, the downloader:

```python
class Downloader:
    def __init__(self, link):
        self.link = link
        
    def download(self):
        # This will download page html
        uClient = uReq(self.link)
        page_html = uClient.read()
        uClient.close()
        return soup(page_html, "lxml")
    
    def get_urls(self):
        #This will scrape urls for characters:
        page_html = self.download()
        lis = page_html.find('span',{'id':'Ankh-Morpork_and_The_Watch'}).parent.previous.previous.parent.parent.parent.find_all('li')
        urls = []
        for li in lis:
            # this connects general part of url t specific
            urls.append('https://wiki.lspace.org' + li.find('a')['href'])   
        return urls
    
    def get_char_names(self):
        # This method is supposed to give list of characters
        page_html = self.download()
        lis = page_html.find('span',{'id':'Ankh-Morpork_and_The_Watch'}).parent.previous.previous.parent.parent.parent.find_all('li')
        names = []
        for li in lis:
            # So I get nice list
            names.append(li.a.text)
        return names
    
    
    def get_book(self):
        # This should get me the book, not done yet
        urls = self.get_urls()
        books = []
        for url in urls:
            #Still not working
            books.append(Downloader(url).download().find('b', {'title':'Bibliography'}.findAll('title').next_sibling.strip())
        return books
        
    def dataf(self):
        #this will create the dataframe, get_book need to be finished to work
        names = self.get_char_names()
        books = self.get_book()
        
        self.data = pd.DataFrame({'name': pd.Series(names),
             'books': pd.Series(books)})
```
